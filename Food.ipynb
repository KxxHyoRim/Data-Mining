```python
#불러올 파일의 경로를 filename 변수에 저장
file1 = '/content/drive/My Drive/Data Mining/food_1.csv'
file2 = '/content/drive/My Drive/Data Mining/food_2.csv'

#pandas read_csv로 불러오기
df1 = pd.read_csv(file1, encoding='cp949') # encoding : 한글 로딩 위함
df1.shape
```

```python
#pandas read_csv로 불러오기
df2 = pd.read_csv(file2, encoding='cp949') # encoding : 한글 로딩 위함
df2.shape
```

## 불필요한 열 제거

```python
df1 = df1.drop(columns=['품목별'])
df2 = df2.drop(columns=['품목별'])
```

## 월별 Column에 index로 접근하기 위해 List를 활용하여 String 생성

```python
year = list(range(2000, 2010+1))
print(year)

month = list(range(1,12+1))
print("Int : ", end='')
print( month)

idx = 0
for m in month:
  if (len(str(m)) < 2):
    month[idx] = '0' + str(m)
  else :
    month[idx] = str(m)
  idx += 1

print("Str : ", end='')
print( month)
```

## For 문을 활용하여 지역별로 연간 짜장, 짬뽕 데이터를 시각화

```python
titles = ['Nationwide', 'Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan']


for local in range(8):
  jajang = []
  jjambbong = []

  for y in year:
    sum1 = 0
    sum2 = 0

    for m in month:
      col_idx = str(y) + ". " + m   # ex) '2000. 01'
      sum1 += df1[col_idx][local]
      sum2 += df2[col_idx][local]
    jajang.append(sum1/12)
    jjambbong.append(sum2/12)

  print("\n[" + str(y) + "년]\n")
  print('짜장면', end = "  ")
  print(jajang)
  print('짬뽕', end = "  ")
  print(jjambbong)

  fig = plt.figure( figsize=(6, 3))
  ax = fig.add_subplot()
  ax.plot(jajang, label='jajang')
  ax.plot(jjambbong, label='jjambong')
  ax.legend()
  ax.set_title(titles[local])
  plt.show()
```
